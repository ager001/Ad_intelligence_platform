```python
# 🟢 STEP 1: IMPORT + LOAD
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Projects\ML_PROJECTS\Ad_intelligence_platform\data\raw\ad_click.csv")

df.head()
```

```python
# 🟢 STEP 2: CHECK MISSING VALUES (REAL VIEW)
df.isnull().sum()
```

#🧠 STEP 3: CLEANING STRATEGY (VERY IMPORTANT)
We decide scientifically:

Column	Action
age	fill with median
browsing_history	drop
gender	fill "Unknown"
device_type	fill "Unknown"
ad_position	fill "Unknown"
time_of_day	fill "Unknown"

```python
#🧹 STEP 4: DROP HEAVY MISSING COLUMN
df.drop(columns=["browsing_history"], inplace=True, errors="ignore")
```

```python
# STEP 5: HANDLE AGE (NUMERIC IMPUTATION)
df["age"] = df["age"].fillna(df["age"].median())
```

```python
# STEP 6: HANDLE CATEGORICAL MISSING VALUES
df["gender"] = df["gender"].fillna("Unknown")
df["device_type"] = df["device_type"].fillna("Unknown")
df["ad_position"] = df["ad_position"].fillna("Unknown")
df["time_of_day"] = df["time_of_day"].fillna("Unknown")
```

```python
# 🧪 STEP 7: VERIFY CLEANING

df.isnull().sum()
```

```python
# STEP 8: CHECK DATA TYPES
df.info()
```

```python
# 🧹 STEP 9: STANDARDIZE CATEGORICAL DATA
categorical_cols = ["gender", "device_type", "ad_position", "time_of_day"]

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```
