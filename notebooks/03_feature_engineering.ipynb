```python
# Importing all necessary libararies
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/ad_click_clean.csv")

df.head()
```

```python
# Here we identify feature types that we have in our dataset
df.info()
```

```python
#In this cell we drop non-predictive columns 
df.drop(columns=["id", "full_name"], inplace=True)
```

```python
# Encoding target variables
df["click"].unique()
```

```python
#One Hot Encoding 
df_encoded = pd.get_dummies(
    df,
    columns=["gender", "device_type", "ad_position", "time_of_day"],
    drop_first=True
)



df_encoded = df_encoded.astype({
    col: 'int'
    for col in df_encoded.columns
    if df_encoded[col].dtype == 'bool'
})

df_encoded.head()
```

```python
df_encoded.drop(columns=["id", "full_name"], inplace=True, errors="ignore")
```
